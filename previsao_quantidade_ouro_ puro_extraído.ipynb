## Calculando sMAPE

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from functools import partial
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import make_scorer
import numpy as np
```

```python
# As características são todas as informações sobre os clientes que não possuem quaisquer relações com sua identificação:
features_test = df_test_new_dropna.drop(['rougher.output.recovery', 'final.output.recovery','date'], axis=1)

# O resultado que queremos:
target_test_1 = df_test_new_dropna['rougher.output.recovery']
target_test_2 = df_test_new_dropna['final.output.recovery']
```

```python
# As características são todas as informações sobre os clientes que não possuem quaisquer relações com sua identificação:
features_train = df_train_new_dropna.drop(['rougher.output.recovery', 'final.output.recovery','date'], axis=1)

# O resultado que queremos:
target_train_1 = df_train_new_dropna['rougher.output.recovery']
target_train_2 = df_train_new_dropna['final.output.recovery']
```

```python
# Criando a função para metrificar nossos modelos:

def smape(teste, predição):
    denominador = (np.abs(teste) + np.abs(predição))/2
    numerador = np.abs(teste - predição)
    soma = np.sum(np.divide(numerador, denominador, out=np.zeros_like(denominador), where=denominador!=0))
    smape = (soma / len(teste)) * 100
    return smape


# As características são todas as informações sobre os clientes que não possuem quaisquer relações com sua identificação:
features_train = df_train_new_dropna.drop(['rougher.output.recovery', 'final.output.recovery','date'], axis=1)

# O resultado que queremos:
target_train_1 = df_train_new_dropna['rougher.output.recovery']
target_train_2 = df_train_new_dropna['final.output.recovery']

# As características são todas as informações sobre os clientes que não possuem quaisquer relações com sua identificação:
features_test = df_test_new_dropna.drop(['rougher.output.recovery', 'final.output.recovery','date'], axis=1)

# O resultado que queremos:
target_test_1 = df_test_new_dropna['rougher.output.recovery']
target_test_2 = df_test_new_dropna['final.output.recovery']

# ajustando a escalabilidade das características, para não haver uma distinção de importância nas características do modelo.

numeric_train = df_train_new_dropna.drop(['date','final.output.recovery', 'rougher.output.recovery'], axis =1)
numeric_train = numeric_train.columns
numeric_train

scaler = StandardScaler()
scaler.fit(features_train[numeric_train])
features_train[numeric_train] = scaler.transform(features_train[numeric_train])

numeric_test = df_test_new_dropna.drop(['date','final.output.recovery', 'rougher.output.recovery'], axis =1)
numeric_test = numeric_test.columns
numeric_test

scaler = StandardScaler()
scaler.fit(features_train[numeric_test])
features_test[numeric_test] = scaler.transform(features_test[numeric_test])

#Definindo os modelos
lr = LinearRegression()
dt = DecisionTreeRegressor(random_state=12345)
rf = RandomForestRegressor(n_estimators=20, random_state=12345)

# Define o scorer SMAPE
smape_scorer = make_scorer(smape, greater_is_better=False)

# Define o KFold com 5 folds
kf = KFold(n_splits=5, shuffle=True, random_state=12345)

# Avalia os modelos usando KFold e SMAPE
scores_lr1 = -cross_val_score(lr, features_train, target_train_1, cv=kf, scoring=smape_scorer)
scores_lr2 = -cross_val_score(lr, features_train, target_train_2, cv=kf, scoring=smape_scorer)
scores_dt1 = -cross_val_score(dt, features_train, target_train_1, cv=kf, scoring=smape_scorer)
scores_dt2 = -cross_val_score(dt, features_train, target_train_2, cv=kf, scoring=smape_scorer)
scores_rf1 = -cross_val_score(rf, features_train, target_train_1, cv=kf, scoring=smape_scorer)
scores_rf2 = -cross_val_score(rf, features_train, target_train_2, cv=kf, scoring=smape_scorer)

print('Regressão Linear:')
print(f'SMAPE Alvo 1: {scores_lr1.mean():.2f} +/- {scores_lr1.std():.2f}')
print(f'SMAPE Alvo 2: {scores_lr2.mean():.2f} +/- {scores_lr2.std():.2f}')
print('Árvore de Decisão:')
print(f'SMAPE Alvo 1: {scores_dt1.mean():.2f} +/- {scores_dt1.std():.2f}')
print(f'SMAPE Alvo 2: {scores_dt2.mean():.2f} +/- {scores_dt2.std():.2f}')
print('Floresta Aleatória:')
print(f'SMAPE Alvo 1: {scores_rf1.mean():.2f} +/- {scores_rf1.std():.2f}')
print(f'SMAPE Alvo 2: {scores_rf2.mean():.2f} +/- {scores_rf2.std():.2f}')
```

O modelo Regressão Floresta Aleatória é o que apresenta a menor variação média do SMAPE entre os modelos que treinamos. Por isso vou utilizar este modelo para o teste.

### Regressão da Floresta Aleatória

```python
modelo_teste = RandomForestRegressor(n_estimators=20, random_state=12345)
modelo_teste.fit(features_train,target_train_1) 
predictions_valid_1 = modelo_teste.predict(features_test)

modelo_teste.fit(features_train,target_train_2)
predictions_valid_2 = modelo_teste.predict(features_test)
```

```python
smape_r = smape(target_test_1, predictions_valid_1)
smape_f = smape(target_test_2, predictions_valid_2)
```

```python
smape_final = (0.25*smape_r)+(0.75*smape_f)
smape_final
```

Primeiro fizemos o pré-processamento dos dados e deixamos eles prontos para serem trabalhados. Após isso fizemos uma avaliação do comportamento dos resultados durante as fases do procedimento. Usamos a validação cruzada para verificar qual o melhor modelo para nosso projeto e, com base no smape, foi a Regressão da Floresta Aleatória. Depois de decidir qual modelo deveríamos usar, aplicamos o conjunto no conjunto de teste e comparamos os resultados com a métrica que nos foi fornecida, o smape_final que utiliza o smape(rougher) e smape(final), poir isso decidi separar ambas predições. O no final encontramos que existe uma diferença de aproximandamente 14.5 entre o conjunto de predição e o de teste.
